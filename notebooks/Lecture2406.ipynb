```python
import pandas as pd
import matplotlib as px
import matplotlib.pyplot as py
%matplotlib inline
```

# Data Set Loading for Visualisation
Our World In Data look for the link with .org

```python
url="https://covid.ourworldindata.org/data/owid-covid-data.csv"
```

```python
df_country_info=pd.read_csv(url,sep=',')
```

```python
df_country_info.columns
df_country_info.head()
df_country_info.to_csv('covid_full_data.csv', ';')
```

```python
df_country_info['location'].unique()
df_country_info['location']=='Germany'
df_GER=df_country_info[df_country_info['location']=='Germany']
df_GER.reset_index(drop=True)
```

```python
df_country_info['location'].unique()
df_country_info['location']=='United Kingdom'
df_UK=df_country_info[df_country_info['location']=='United Kingdom']
df_UK.reset_index(drop=True)
```

```python
df_country_info['location'].unique()
df_country_info['location']=='Hong Kong'
df_HKG=df_country_info[df_country_info['location']=='Hong Kong']
df_HKG.reset_index(drop=True)
```

```python
df_GER.plot(kind='line', x="date", y="total_cases")
py.show()
```

```python
df_UK.plot(kind='line', x="date", y="total_cases")
py.show()
```

```python
df_HKG.plot(kind='line', x="date", y="total_cases")
py.show()
```

```python
df_ratio=df_GER['total_cases']/df_GER['population']
df_fin=pd.concat([df_GER, df_ratio])
df_fin.rename(columns={67:'cases_pop_ratio'})
```

```python
df_fin.plot(kind='line', x='date', y=0)

py.show()
```
